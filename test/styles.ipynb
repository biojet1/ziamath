```python
import ziamath as zm
font = None
zm.config.svg2 = False
```

## Mathml styles/colors

```python
# mathcolor is inherited from parent
zm.Math('''
<mfrac mathcolor="red">
    <mrow>
        <mi mathcolor="blue">a</mi>
        <mo>+</mo>
        <mn>1</mn>
    </mrow>
    <mn>2</mn>
</mfrac>
''', font=font)
```

```python
# mathvariants
zm.Math('''
<mfrac mathvariant="script">
    <mrow>
        <mi mathcolor="blue">a</mi>
        <mo>+</mo>
        <mi>b</mi>
    </mrow>
    <mi mathvariant="bold">c</mi>
</mfrac>
''', font=font)
```

```python
zm.Math('''<math xmlns="http://www.w3.org/1998/Math/MathML">
  <msup>
    <mi>e</mi>
    <mrow mathcolor='#cc0000'>
      <mn>2</mn>
      <mi>x</mi>
      <mo>+</mo>
      <mn>1</mn>
    </mrow>
  </msup>
</math>''', font=font)
```

```python
zm.Math.fromlatex(r'\sqrt{a^2 + b^2}', color='blue', mathstyle='script', font=font)
```

```python
zm.Math('''
<mfrac>
    <mrow mathbackground="yellow">
        <mi>a</mi>
        <mo>+</mo>
        <mi>b</mi>
    </mrow>
    <mi>c</mi>
</mfrac>
''', font=font)
```

```python
zm.Math('''
<mfrac mathbackground="#b5c3eb">
    <mrow>
        <mi>a</mi>
        <mo>+</mo>
        <mi>b</mi>
    </mrow>
    <mi>c</mi>
</mfrac>
''', font=font)
```

```python
import urllib.request
import os
```

```python
# Pull some math-fonts from CTAN to test
def testfont(exprs, fonturl):
    name = os.path.split(fonturl)[1]
    if not os.path.exists(name):
        urllib.request.urlretrieve(str(fonturl), name)
    
    for expr in exprs:
        display(zm.Latex(expr, font=name))
        print()

testexprs = [
    r'\sqrt{a^2 + b^2}',
    r"x' + x^\prime + x\prime + x^\dprime + x^\trprime + x^\backprime",
    r'\frac{1 + \alpha}{1 + \beta} \frac{1+x}{1-x}',
    r'x = \left(\begin{matrix}a & b\\c & d\end{matrix}\right)',
    r'(x) + \left(x^2\right) + \left(x_0\right)',
    r'\left( \frac{\frac{1}{2}}{x+1} \right)^2',
    r'\overline{1} + \underline{1} + \overline{ABC} + \underline{ABC}'
]
```

```python
testfont(testexprs, 'http://mirrors.ctan.org/fonts/libertinus-fonts/otf/LibertinusMath-Regular.otf')
```

```python
testfont(testexprs, 'http://mirrors.ctan.org/fonts/firamath/FiraMath-Regular.otf')
```

```python
testfont(testexprs, 'http://mirrors.ctan.org/fonts/erewhon-math/Erewhon-Math.otf')
```

```python
testfont(testexprs, 'http://mirrors.ctan.org/fonts/concmath-otf/Concrete-Math.otf')
```
