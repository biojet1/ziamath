```python
import ziamath as zm
font = None
```

```python
# mathcolor is inherited from parent
zm.Math('''
<mfrac mathcolor="red">
    <mrow>
        <mi mathcolor="blue">a</mi>
        <mo>+</mo>
        <mn>1</mn>
    </mrow>
    <mn>2</mn>
</mfrac>
''', font=font)
```

```python
# mathvariants
zm.Math('''
<mfrac mathvariant="script">
    <mrow>
        <mi mathcolor="blue">a</mi>
        <mo>+</mo>
        <mi>b</mi>
    </mrow>
    <mi mathvariant="bold">c</mi>
</mfrac>
''', font=font)
```

```python
zm.Math('''<math xmlns="http://www.w3.org/1998/Math/MathML">
  <msup>
    <mi>e</mi>
    <mrow mathcolor='#cc0000'>
      <mn>2</mn>
      <mi>x</mi>
      <mo>+</mo>
      <mn>1</mn>
    </mrow>
  </msup>
</math>''', font=font)
```

```python
zm.Math.fromlatex(r'\sqrt{a^2 + b^2}', color='blue', mathstyle='script', font=font)
```

```python
zm.Math('''
<mfrac>
    <mrow mathbackground="yellow">
        <mi>a</mi>
        <mo>+</mo>
        <mi>b</mi>
    </mrow>
    <mi>c</mi>
</mfrac>
''', font=font)
```
