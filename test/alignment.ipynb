```python
import xml.etree.ElementTree as ET
from IPython.display import SVG
```

```python
# Rotations - default mode
# Blue dot (origin) should be on red square (ortho bounding box)
# (See https://matplotlib.org/stable/gallery/text_labels_and_annotations/demo_text_rotation_mode.html)
svg = ET.Element('svg')
svg.attrib['xmlns'] = 'http://www.w3.org/2000/svg'
svg.attrib['xmlns:xlink'] = 'http://www.w3.org/1999/xlink'
svg.attrib['width'] = '500'
svg.attrib['height'] = '200'
svg.attrib['viewBox'] = '0 0 500 200'

for xy, halign, valign in [((20, 20), 'left', 'top'),
                           ((100, 20), 'center', 'top'),
                           ((180, 20), 'right', 'top'),
                           ((20, 80), 'left', 'center'),
                           ((100, 80), 'center', 'center'),
                           ((180, 80), 'right', 'center'),
                           ((20, 130), 'left', 'base'),
                           ((100, 120), 'center', 'base'),
                           ((180, 110), 'right', 'base'),
                           ((20, 180), 'left', 'bottom'),
                           ((100, 180), 'center', 'bottom'),
                           ((180, 180), 'right', 'bottom'),

                          ]:
    circ = ET.SubElement(svg, 'circle')
    circ.set('cx', str(xy[0]))
    circ.set('cy', str(xy[1]))
    circ.set('r', '3')
    circ.set('stroke', 'blue')
    circ.set('fill', 'blue')

    zm.Text('$\\sqrt{a^2}$\nABC', rotation=30,
                 size=12,
                 rotation_mode='default',
                 halign=halign, valign=valign).drawon(svg, xy[0], xy[1])

SVG(ET.tostring(svg))
```

```python
# Rotations - anchor mode
# Blue dot (origin) should be on blue square (tight/rotated bounding box)
# (See https://matplotlib.org/stable/gallery/text_labels_and_annotations/demo_text_rotation_mode.html)
svg = ET.Element('svg')
svg.attrib['xmlns'] = 'http://www.w3.org/2000/svg'
svg.attrib['xmlns:xlink'] = 'http://www.w3.org/1999/xlink'
svg.attrib['width'] = '500'
svg.attrib['height'] = '200'
svg.attrib['viewBox'] = '0 0 500 200'

for xy, halign, valign in [((20, 20), 'left', 'top'),
                           ((100, 20), 'center', 'top'),
                           ((180, 20), 'right', 'top'),
                           ((20, 80), 'left', 'center'),
                           ((100, 80), 'center', 'center'),
                           ((180, 80), 'right', 'center'),
                           ((20, 130), 'left', 'base'),
                           ((100, 120), 'center', 'base'),
                           ((180, 110), 'right', 'base'),
                           ((20, 180), 'left', 'bottom'),
                           ((100, 180), 'center', 'bottom'),
                           ((180, 180), 'right', 'bottom'),

                          ]:
    circ = ET.SubElement(svg, 'circle')
    circ.set('cx', str(xy[0]))
    circ.set('cy', str(xy[1]))
    circ.set('r', '3')
    circ.set('stroke', 'blue')
    circ.set('fill', 'blue')

    zm.Text('$\\sqrt{a^2}$\nABC', rotation=30,
                 size=12,
                 rotation_mode='anchor',
                 halign=halign, valign=valign).drawon(svg, xy[0], xy[1])

SVG(ET.tostring(svg))
```

## Alignments

```python
zm.config.debug = True

svg = ET.Element('svg')
svg.attrib['width'] = '500'
svg.attrib['height'] = '200'
svg.attrib['xmlns'] = 'http://www.w3.org/2000/svg'
svg.attrib['viewBox'] = f'0 0 500 200'

def circle(x, y):
    circ = ET.SubElement(svg, 'circle')
    circ.attrib['cx'] = str(x)
    circ.attrib['cy'] = str(y)
    circ.attrib['r'] = '3'
    circ.attrib['fill'] = 'red'
    circ.attrib['stroke'] = 'red'

def testalign(x, y, i=0, halign='left', valign='baseline'):
    circle(x, y)
    expr = r'z = \frac{' + str(i) + '}{2}'
    zm.Latex(expr).drawon(svg, x, y, halign=halign, valign=valign)

x, y = 100, 70
testalign(20, 50, i=0, valign='baseline')
testalign(100, 50, i=1, valign='bottom')
testalign(180, 50, i=2, valign='center')
testalign(260, 50, i=3, valign='axis')
testalign(340, 50, i=4, valign='top')
testalign(20, 120, i=5, halign='left', valign='bottom')
testalign(150, 120, i=6, halign='center', valign='center')
testalign(280, 120, i=7, halign='right', valign='top')

SVG(ET.tostring(svg, encoding='unicode'))
```
