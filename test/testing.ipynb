```python
import ziamath as zm
```

```python
# Alignment of fractions and enclosed fractions
s = '''
    <math>
      <mrow>
      <mo>(</mo>
        <mfrac linethickness="1">
          <mrow><mn>5</mn></mrow>
          <mrow><mn>2</mn></mrow>
        </mfrac>
      <mo>)</mo>
      </mrow>
      <mo>=</mo><mn>10</mn><mo>=</mo>
        <mfrac linethickness="1">
          <mrow><mn>5</mn></mrow>
          <mrow><mn>2</mn></mrow>
        </mfrac>
    </math>
'''
zm.Math(s)
```

```python
# Fenced operator with nothing in it
s = '''<math display='inline' xmlns='http://www.w3.org/1998/Math/MathML'>
   <msup>
      <mrow>
        <mo>|</mo>
      </mrow>
      <mrow>
         <mn>2</mn>
      </mrow>
   </msup>
   </math>'''
zm.Math(s)
```

```python
# Fenced
zm.Math('''<math>
  <mover>
    <mi>v</mi>
    <mo>&rarr;</mo>
  </mover>
  <mo>=</mo>
  <mfenced>
    <msub>
      <mi>v</mi>
      <mn>1</mn>
    </msub>
    <msub>
      <mi>v</mi>
      <mn>2</mn>
    </msub>
    <msub>
      <mi>v</mi>
      <mn>3</mn>
    </msub>
  </mfenced>
</math>''')
```

```python
zm.Math('''
<mfenced separators=',;:,'>
    <mo>a</mo>
    <mo>b</mo>
    <mo>c</mo>
    <mo>d</mo>
</mfenced>
''')
```

```python
# Comments in the XML, classnames in the tags
s = '''<math display='inline' xmlns='http://www.w3.org/1998/Math/MathML'>
<mi>e</mi>
<mo class='MathClass-rel'>=</mo>
<msub>
    <mrow>
        <mi>E</mi>
    </mrow>
    <mrow>
        <mi>m</mi>
    </mrow>
</msub>
<mi class='qopname'> cos</mi>
<mo> ⁡<!-- FUNCTION APPLICATION --> </mo>
<!-- nolimits -->
<mi mathvariant='italic'>ωt</mi>
</math>
'''
zm.Math(s)
```

```python
# Division sign
s = '''
<math xmlns="http://www.w3.org/1998/Math/MathML">
    <mrow>
        <mn>0</mn>
        <mo>÷</mo>
        <mn>23</mn>
    </mrow>
    <mo>=</mo>
    <mn>0</mn>
</math>'''
zm.Math(s)
```

```python
s = '''<math xmlns="http://www.w3.org/1998/Math/MathML">
            <mrow>
                <mi>x</mi>
                <mo>=</mo>
                <mfrac>
                    <mrow>
                        <mo form="prefix">-</mo>
                        <mi>b</mi>
                        <mo>±<!-- &amp;PlusMinus; --></mo>
                        <msqrt>
                            <msup>
                                <mi>b</mi>
                                <mn>2</mn>
                            </msup>
                            <mo>-</mo>
                            <mn>4</mn>
                            <mi>a</mi>
                            <mi>c</mi>
                        </msqrt>
                    </mrow>
                    <mrow>
                        <mn>2</mn>
                        <mi>a</mi>
                    </mrow>
                </mfrac>
            </mrow>
        </math>'''
zm.Math(s)
```
