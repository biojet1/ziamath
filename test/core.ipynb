```python
import ziamath as zm
font = None
```

```python
# SVG2 mode (not supported by all browsers/viewers)
zm.config.svg2 = True
svg = zm.Text('This is math: $x^2 + y^2$')
assert 'symbol' in svg.svg()
svg
```

```python
# SVG1.x mode doesn't use <symbol> and <use> to place glyphs
zm.config.svg2 = False
svg = zm.Text('This is math: $x^2 + y^2$')
assert 'symbol' not in svg.svg()
svg
```

```python
zm.config.svg2 = True
```

```python
zm.config.minsizefraction = .5
zm.Math.fromlatex('a^{2^{2^{2^2}}}', font=font)
```

```python
zm.config.minsizefraction = .1
zm.Math.fromlatex('a^{2^{2^{2^2}}}', font=font)
```

```python
# SVG precision
zm.config.precision = 3
zm.Math.fromlatex('a', font=font).svg()
```

```python
# SVG precision
zm.config.precision = 1
zm.Math.fromlatex('a', font=font).svg()
```

```python
zm.Latex(r'1 + \blah(x)')   # blah isn't a function
```

```python
zm.declareoperator(r'\blah')
zm.Latex(r'1 + \blah(x)')   # Now it is.
```

```python
# ziamath declares a few trig alternates that latex2mathml doesn't have
zm.Latex(r'1 + \tg\theta + \ctg\theta + \arctg\theta + \arcctg\theta')
```

```python
zm.Latex(r'a+b \kern-1em c \\ a+b \kern-18mu c \\ a+b \kern-1cm c')  # 18mu = 1em; first two should place c in same position
```

```python
# Decimal separator only affects commas between digits
zm.config.decimal_separator = '.'
zm.Latex(r'1,23 + 12\,3 + (x, y, z)')
```

```python
zm.config.decimal_separator = ','
zm.Latex(r'1,23 + 12\,3 + (x, y, z)')
```

```python
import ziafont as zf
import xml.etree.ElementTree as ET
from IPython.display import SVG
```

```python
svg = ET.Element('svg')
svg.attrib['xmlns'] = 'http://www.w3.org/2000/svg'
svg.attrib['xmlns:xlink'] = 'http://www.w3.org/1999/xlink'
svg.attrib['width'] = '300'
svg.attrib['height'] = '200'
svg.attrib['viewBox'] = '0 0 300 200'

zm.Text('ABC', color='orange').drawon(svg, 100, 100)
zm.Math.fromlatex('a^2', color='green').drawon(svg, 100, 180)
zm.Text('Hi $n^2$', color='maroon').drawon(svg, 200, 100)
zf.Text('What', color='yellow').drawon(svg, 180, 180)
SVG(ET.tostring(svg))
```
