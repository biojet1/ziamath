## Inline vs Block mode

Second term with `\limits` should always place limits above/below the operator 

```python
zm.Latex(r'R = \sum_{i=1}^n R_i + \sum\limits_{i=1}^n R_i', inline=False)
```

```python
zm.Latex(r'R = \sum_{i=1}^n R_i + \sum\limits_{i=1}^n R_i', inline=True)
```

```python
zm.Latex(r'v = \frac{1}{C} \int_0^T i \mathrm{d}t + \frac{1}{C} \int\limits_0^T i \mathrm{d}t', inline=False)
```

```python
zm.Latex(r'v = \frac{1}{C} \int_0^T i \mathrm{d}t + \frac{1}{C} \int\limits_0^T i \mathrm{d}t', inline=True)
```

```python
zm.Latex(r'f = \prod_{j=0}^J k_j + \prod\limits_{j=0}^J k_j', inline=True)
```

```python
zm.Latex(r'f = \prod_{j=0}^J k_j + \prod\limits_{j=0}^J k_j', inline=False)
```

```python
# dfrac always big, tfrac always small, and frac depends on inline/block mode.
zm.Latex(r'x\dfrac{1}{2}x + x\tfrac{1}{2}x + x\frac{1}{2}x', inline=True)
```

```python
zm.Latex(r'x\dfrac{1}{2}x + x\tfrac{1}{2}x + x\frac{1}{2}x', inline=False)
```

```python
zm.config.minsizefraction = .5
zm.Latex('''a_0+{1\over a_1+
      {1\over a_2+
        {1 \over a_3 + 
           {1 \over a_4}}}}''', inline=False)
```

```python
zm.Latex('''a_0+{1\over a_1+
      {1\over a_2+
        {1 \over a_3 + 
           {1 \over a_4}}}}''', inline=True)
```

```python
zm.Latex('''a_0+{1\over\displaystyle a_1+
      {1\over\displaystyle a_2+
        {1 \over\displaystyle a_3 + 
           {1 \over\displaystyle a_4}}}}''', inline=False)
```
