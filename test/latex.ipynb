## Latex examples

Select examples from https://en.wikibooks.org/wiki/LaTeX/Mathematics

Note latex2mathml package doesn't support everything.

```python
import ziamath as zm
font = None
```

```python
zm.Math.fromlatex(r'\frac{1 + \alpha}{1 + \beta}', font=font)
```

```python
zm.Math.fromlatex(r'k_{n+1} = n^2 + k_n^2 - k_{n-1}', font=font)
```

```python
zm.Math.fromlatex('f(n) = n^5 + 4n^2 + 2 |_{n=17}', font=font)
```

```python
zm.Math.fromlatex(r'\frac{n!}{k!(n-k)!} = \binom{n}{k}', font=font)
```

```python
zm.Math.fromlatex(r'\frac{\frac{1}{x}+\frac{1}{y}}{y-z}', font=font)
```

```python
zm.Math.fromlatex(r'\sqrt{\frac{a}{b}}', font=font)
```

```python
zm.Math.fromlatex(r'\sum_{i=1}^{10} t_i', font=font)
```

```python
zm.Math.fromlatex(r'\left(\frac{x^2}{y^3}\right)', font=font)
```

```python
zm.Math.fromlatex(r'\frac{\mathrm d}{\mathrm d x} \left( k g(x) \right)', font=font)
```

```python
zm.Math.fromlatex(r'''\begin{matrix}
  a & b & c \\
  d & e & f \\
  g & h & i
 \end{matrix}''', font=font)
```

```python
zm.Math.fromlatex(r'^2', font=font)  # Superscript with no base
```

## Alignments

```python
from IPython.display import SVG
from xml.etree import ElementTree as ET
```

```python
zm.config.debug = True

svg = ET.Element('svg')
svg.attrib['width'] = '500'
svg.attrib['height'] = '200'
svg.attrib['xmlns'] = 'http://www.w3.org/2000/svg'
svg.attrib['viewBox'] = f'0 0 500 200'

def circle(x, y):
    circ = ET.SubElement(svg, 'circle')
    circ.attrib['cx'] = str(x)
    circ.attrib['cy'] = str(y)
    circ.attrib['r'] = '3'
    circ.attrib['fill'] = 'red'
    circ.attrib['stroke'] = 'red'

def testalign(x, y, i=0, halign='left', valign='baseline'):
    circle(x, y)
    expr = r'z = \frac{' + str(i) + '}{2}'
    zm.Math.fromlatex(expr).drawon(svg, x, y, halign=halign, valign=valign)

x, y = 100, 70
testalign(20, 50, i=0, valign='baseline')
testalign(100, 50, i=1, valign='bottom')
testalign(180, 50, i=2, valign='center')
testalign(260, 50, i=3, valign='axis')
testalign(340, 50, i=4, valign='top')
testalign(20, 120, i=5, halign='left', valign='bottom')
testalign(150, 120, i=6, halign='center', valign='center')
testalign(280, 120, i=7, halign='right', valign='top')

SVG(ET.tostring(svg, encoding='unicode'))
```

```python
zm.config.debug = True

svg = ET.Element('svg')
svg.attrib['width'] = '500'
svg.attrib['height'] = '200'
svg.attrib['xmlns'] = 'http://www.w3.org/2000/svg'
svg.attrib['viewBox'] = f'0 0 500 200'

def circle(x, y):
    circ = ET.SubElement(svg, 'circle')
    circ.attrib['cx'] = str(x)
    circ.attrib['cy'] = str(y)
    circ.attrib['r'] = '3'
    circ.attrib['fill'] = 'red'
    circ.attrib['stroke'] = 'red'

def testalign(x, y, i=0, halign='left', valign='baseline'):
    circle(x, y)
    expr = f'Hi $a^{i}$'
    zm.Text(expr, mathfont=font).drawon(svg, x, y, halign=halign, valign=valign)

x, y = 100, 70
testalign(20, 50, i=0, valign='baseline')
testalign(100, 50, i=1, valign='bottom')
testalign(180, 50, i=2, valign='center')
testalign(280, 50, i=3, valign='top')
testalign(20, 120, i=4, halign='left', valign='bottom')
testalign(150, 120, i=5, halign='center', valign='center')
testalign(280, 120, i=6, halign='right', valign='top')

SVG(ET.tostring(svg, encoding='unicode'))
```
